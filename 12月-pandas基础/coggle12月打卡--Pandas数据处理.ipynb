## 任务4：Pandas数据分组聚合
任务要点：groupby、agg、transform
- 步骤1：读取文件https://cdn.coggle.club/Pokemon.csv
- 步骤2：学习groupby分组聚合的使用
- 步骤3：学习agg分组聚合的使用
- 步骤4：学习transform的使用
- 步骤5：使用groupby、agg、transform，统计数据在Type 1分组下 HP的均值

```python
df4 = pd.read_csv('https://cdn.coggle.club/Pokemon.csv')
```

```python
df4.head()
```

```python
# 按  'Type 1','Type 2' 分组后，对Total求和

df4.groupby(['Type 1','Type 2'])['Total'].sum()
```

```python
# 用agg方法，统计聚合后多个统计量
df4.groupby(['Type 1','Type 2'])['Total'].agg(['min', 'mean', 'max'])

# 还可以定义一个字典

agg_func ={'Total':['min', 'mean', 'max']}
df4.groupby(['Type 1','Type 2']).agg(agg_func)
```

transform的使用场景：
![20211214115054](https://cdn.jsdelivr.net/gh/xihuishawpy/PicBad@main/blogs/pictures/20211214115054.png)

分析统计出来的数据需要与原始数据加工到一起，但新旧数据的长度是不一样的
1. 常规做法是merge；
2. 用transform会更方便；

```python
# merge 

total_hp = df4.groupby(['Type 1'])['HP'].sum().rename('total_hp').reset_index()

df4_1 = df4.merge(total_hp)
df4_1['HP_ratio']=df4_1['HP']/df4_1['total_hp']

df4_1
```

```python
# transform

df4['HP_ratio']=df4['HP']/df4.groupby(['Type 1'])['HP'].transform('sum')
df4
```

## 任务5：Pandas日期数据处理
任务要点：日期处理
- 步骤1：创建一列dt，dt取值为从1638263656 到 1638283656 的 unix时间
- 步骤2：将dt列转为datatime格式
- 步骤3：筛选出dt列中小时为10的行
- 步骤4：将dt列整体增加8小时的时间

```python
dt = [d for d in range(1638263656,1638283657)]

dt_df = pd.DataFrame(pd.to_datetime(dt,unit='s'),columns=['dt'])

dt_df.head()
```

```python
dt_df[dt_df['dt'].dt.hour==10]
```

```python
dt_df['dt_plus8h'] = dt_df['dt']+pd.Timedelta(8,unit='hours')
dt_df
```
